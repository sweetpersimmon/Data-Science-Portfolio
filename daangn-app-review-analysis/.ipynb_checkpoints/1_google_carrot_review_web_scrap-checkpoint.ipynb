```python
from selenium.webdriver import Chrome
import time

import requests
import lxml.html
from selenium.webdriver.chrome.options import Options

import numpy as np
import pandas as pd
import re
import pickle
```

# 데이터 수집
selenium을 사용하여 구글 앱스토어에서 당근마켓 어플 리뷰와 별점을 크롤링

```python
url = 'https://play.google.com/store/apps/details?id=com.towneers.www&hl=ko&showAllReviews=true'
```

```python
# 동적 페이지를 크롤링 해야 하기 때문에 
driver = Chrome()
driver.get(url)
driver.maximize_window()
```

```python
# 1. window.scrollTo : 스크롤 바를 끝까지 내려서 모든 리뷰를 표시
# 2. 스크롤 바가 끝까지 내려간 후 더보기 버튼을 클릭하여 추가로 리뷰 표시

et = 0
st = 0
while(et<20 and st<25):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)
    try:
        more=driver.find_element_by_xpath('//span[@class="RveJvd snByac"]')
        if (more is not None):
            more.click()
            st+=1
            et=0
    except:
        et+=1 # 에러 발생 시 et 변수에 1 추가
    
    print(st) # 더보기 버튼을 클릭한 횟수 표시
```

```python
# 3. 텍스트 내용이 길어서 요약된 경우, '자세히 보기'를 눌러 모든 내용을 표시
more1=driver.find_elements_by_css_selector('div.cQj82c button.LkLjZd.ScJHi.OzU4dc  ')
for m in more1:
    try:
        m.click()
    except:
        pass
```

```python
# 리뷰 텍스트 : 한글, 숫자를 제외한 영문/특수문자 제거
# 별점 : 별점 개수 count

data=[]
gra = driver.find_elements_by_css_selector('div.pf5lIe')
rv = driver.find_elements_by_css_selector('div.UD7Dzf')
for r, g in zip(rv, gra):
    grade = len(g.find_elements_by_css_selector('div.vQHuPe.bUWb7c'))
    review = r.find_element_by_css_selector('div.UD7Dzf span').text
    review = review.strip()
    reivew = re.sub(r'\s+',' ',review)
    review = re.sub(r'[^ ㄱ-ㅣ가-힣0-9]+', ' ',review)
    data.append((review,grade))
```

# 데이터 전처리
1. 긍/부정 데이터 구분 후 수동 라벨링
2. 긍/부정 데이터 간의 비율 조정

```python
# 중간 작업을 이어서 하기 위해 df를 pickle 형태로 임시 저장
with open('../data/data.pkl', 'wb') as f:
    pickle.dump(data, f)
```

```python
이어서 불러오기
data = pd.read_pickle('../data/data.pkl')
df = pd.DataFrame(data).rename(columns={0:'document',1:'sentiment'})

# 4점 이상일 때만 긍정(1), 이외에는 부정(0)으로 처리
df['sentiment'] = np.where(df['sentiment']>=4,1,0)

# 리뷰 내용에 공백만 포함되어 있는 경우, 해당 리뷰 제거
df['document'] = df['document'].str.strip()
df['document'] = df['document'].replace('', np.nan)
df = df.dropna(axis=0).reset_index(drop=True)
```

```python
df.groupby('sentiment').size()
```

```python
# sentiment가 1인 경우의 비율이 너무 많기 때문에 랜덤 샘플링으로 감소
df0 = df.query('sentiment==1').reset_index(drop=True)
df0.drop(df0.sample(frac=.45).index, inplace=True)

xdf = pd.concat([df.query('sentiment==0').reset_index(drop=True),df0])
```

```python
xdf.drop(columns='grade', inplace=True)
```

```python
xdf.to_csv('../data/carrot_review.csv',encoding='cp949',index=False)
```
