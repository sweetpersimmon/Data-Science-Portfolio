```python
tmp1 = register.groupby(['year','MCT_CAT_NM'])[['USE_CNT','USE_AMT']].sum().reset_index().query('year==2019')
tmp2 = register.groupby(['year','MCT_CAT_NM'])[['USE_CNT','USE_AMT']].sum().reset_index().query('year==2020')
```

```python
fig, ax = plt.subplots(figsize=(20,20))

x1 = np.log(tmp1.USE_CNT).apply(lambda x : float(format(x, '.2f'))).values
y1 = np.log(tmp1.USE_AMT).apply(lambda x : float(format(x, '.2f'))).values

x2 = np.log(tmp2.USE_CNT).apply(lambda x : float(format(x, '.2f'))).values
y2 = np.log(tmp2.USE_AMT).apply(lambda x : float(format(x, '.2f'))).values

sub_label1 = tmp1.eval('USE_AMT/USE_CNT').values
sub_label2 = tmp2.eval('USE_AMT/USE_CNT').values

ax.scatter(x1,y1, label = '2019')
ax.scatter(x2,y2, color='blue', alpha=.6, label='2020')



for i, (label, sub_label) in enumerate(zip(tmp1.MCT_CAT_NM.tolist(), sub_label1)):
    ax.annotate(f'{label}({sub_label:.2f})',(x1[i]-0.15,y1[i]+0.08), fontsize=9)

    
for i, (label, sub_label) in enumerate(zip(tmp2.MCT_CAT_NM.tolist(), sub_label2)):
    ax.annotate(f'{label}({sub_label:.2f})',(x2[i]-0.15,y2[i]+0.08), fontsize=9)

ax.set_xlabel('총소비횟수_log변환', fontsize=12)
ax.set_ylabel('총소비금액_log변환', fontsize=12)

ax.legend(loc='upper left')

ax.set_title('2019 vs 2020 업종별 총소비현황', fontsize=16)
```

### 시기별 객단가 비교

```python
# 코로나 관련 기간 정의
normal_case = pd.date_range(start='2019-02-01', end='2020-02-20') # 코로나 이전 시기
covid_spread = pd.date_range(start='2020-02-21', end='2020-03-15') # 코로나 확산시기 / 하루 확진자 100명 이상
covid_ease = pd.date_range(start='2020-03-22', end='2020-05-05') #코로나 확산 1주일 후 ~ 생활속 거리두기 시행전
covid_resilence = pd.date_range(start='2020-05-06', end='2020-05-31') # 생활속 거리두기 시행
```

```python
register_normal = register.loc[(register.STD_DD.isin(normal_case)) & (register.MCT_CAT_NM.isin(['의복','신변잡화']))]
register_spread = register.loc[(register.STD_DD.isin(covid_spread)) & (register.MCT_CAT_NM.isin(['의복','신변잡화']))]
register_ease = register.loc[(register.STD_DD.isin(covid_ease)) & (register.MCT_CAT_NM.isin(['의복','신변잡화']))]
register_resilience = register.loc[(register.STD_DD.isin(covid_resilence)) & (register.MCT_CAT_NM.isin(['의복','신변잡화']))]
```

```python
register_normal=register_normal.eval('USE_AMT_PER_CNT = USE_AMT/USE_CNT')
register_spread=register_spread.eval('USE_AMT_PER_CNT = USE_AMT/USE_CNT')
register_ease=register_ease.eval('USE_AMT_PER_CNT = USE_AMT/USE_CNT')
register_resilience=register_resilience.eval('USE_AMT_PER_CNT = USE_AMT/USE_CNT')
```

```python
def detect_outlier(df, col):
    outlier_indices = []
    for dong in df.HDONG_NM.unique():
        Q1,Q3 = df.loc[df.HDONG_NM==dong,col].quantile([.25,.75]).values
        IQR = Q3-Q1

        outlier_indices.extend(df.loc[(df.HDONG_NM==dong) & ((df[col] < Q1-1.5*IQR) | (df[col] > Q3+1.5*IQR))].index)
    return outlier_indices
```
