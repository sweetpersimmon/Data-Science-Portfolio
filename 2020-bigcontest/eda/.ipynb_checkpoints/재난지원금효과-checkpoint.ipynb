```python
import pandas as pd
import numpy as np
import scipy 

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from random import choices,sample

import os
import gc
import pickle

import warnings
warnings.filterwarnings('ignore')

from datetime import timedelta

plt.style.use('ggplot')

mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams["font.family"] = "Malgun Gothic"

os.chdir("../")
root_path = os.getcwd()

data_folder_path = os.path.join(root_path, 'data')
original_file_path = os.path.join(data_folder_path, 'original_data')
original_raw_file_path = os.path.join(original_file_path,'raw_data')
original_processed_file_path = os.path.join(original_file_path,'processed_data')

raw_file_folders = os.listdir(original_raw_file_path)

external_file_path = os.path.join(data_folder_path,'external_data')
external_raw_file_path = os.path.join(external_file_path,'raw_data')
external_processed_file_path = os.path.join(external_file_path,'processed_data')
```

```python
card_files = [file for file in os.listdir(original_processed_file_path) if file.startswith('카드')]

df_list = []

for file_name in card_files:
    file_path = os.path.join(original_processed_file_path, file_name)
    df = pd.read_csv(file_path)
    df_list.append(df)

foreigner = df_list[1]
register = df_list[0]

with open(os.path.join(original_processed_file_path,'card_code_to_name_dict.pickle'), 'rb') as f:
    card_code_to_dong_dict = pickle.load(f)

cat_dict = {10 : '숙박',
20 : '레저용품',
21 : '레저업소',
22 : '문화취미',
30 : '가구',
31 : '전기',
32 : '주방용구',
33 : '연료판매',
34 : '광학제품',
35 : '가전',
40 : '유통업',
42 : '의복',
43 : '직물',
44 : '신변잡화',
50 : '서적문구',
52 : '사무통신',
60 : '자동차판매',
62 : '자동차 정비',
70 : '의료기관',
71 : '보건위생',
80 : '요식업소',
81 : '음료식품',
92 : '수리서비스'}

register['region_code'] = register.GU_CD.astype(str)+register.DONG_CD.astype(str)
register['GU_NM'] = register.region_code.map(card_code_to_dong_dict.get('구명'))
register['HDONG_NM'] = register.region_code.map(card_code_to_dong_dict.get('행정동명'))

register = register.drop('region_code', axis=1)

register['MCT_CAT_NM'] = register.MCT_CAT_CD.map(cat_dict)

register['STD_DD'] = pd.to_datetime(register['STD_DD'], format='%Y%m%d')

register['year'] = register['STD_DD'].dt.year.values
register['month'] = register['STD_DD'].dt.month.values
register['weekofyear'] = register['STD_DD'].dt.weekofyear.values
register['dayofweek'] = register['STD_DD'].dt.dayofweek.values

register['weekday'] = np.where(register.dayofweek<5,'평일','주말')
```

```python
for cat in register.MCT_CAT_NM.unique()[:10]:
    register.loc[register.MCT_CAT_NM==cat].query('year==2020').groupby(['month'])['USE_CNT'].sum().plot(figsize=(8,3), kind='bar', width=1, color='skyblue', alpha=.8)
    plt.title(f'{cat}')
    plt.show()
```

* 거의 모든 업종에서 5월 소비횟수가 늘어남
* 생활속거리두기 등 사회적분위기로 인해 소비가 회복된 것인지, 재난지원금으로 인해 소비가 늘어난 것인지 확인할 방법이 필요

#### 연령별 

```python
for cat in card_consume_in_COVID.MCT_CAT_NM.unique():
    tmp_COVID = card_consume_in_COVID.loc[card_consume_in_COVID.MCT_CAT_NM==cat]
    tmp_DAF = card_consume_in_DAF.loc[card_consume_in_DAF.MCT_CAT_NM==cat]
    
    DAF_AGE=(tmp_DAF.groupby(['AGE_CD'])['USE_CNT'].sum()/tmp_DAF.groupby(['AGE_CD'])['USE_CNT'].sum().sum()).values
    COVID_AGE=(tmp_COVID.groupby(['AGE_CD'])['USE_CNT'].sum()/tmp_COVID.groupby(['AGE_CD'])['USE_CNT'].sum().sum()).values

    indices = np.array(range(len(DAF_AGE)))
    
    plt.figure(figsize=(12,4))

    plt.bar(indices-0.15,COVID_AGE,width=.3, color='b', label='코로나기간')
    plt.bar(indices+0.15,DAF_AGE,width=.3, color='r', label='재난지원금기간')

    plt.xticks(ticks=indices, labels=card_consume_in_DAF.loc[card_consume_in_DAF.MCT_CAT_NM=='숙박'].groupby(['AGE_CD'])['USE_CNT'].sum().index)
    plt.xlabel('나이대')
    plt.ylabel('구매횟수비율')
    plt.legend()
    plt.title(f'{cat}')
    
    plt.show()
```

```python
for cat in card_consume_in_COVID.MCT_CAT_NM.unique():
    fig, ax = plt.subplots(figsize=(18,4), ncols=2)
    sns.barplot(
        data = card_consume_in_COVID.loc[card_consume_in_COVID.MCT_CAT_NM==cat].groupby(['AGE_CD','SEX_CD'])['USE_CNT'].sum().reset_index(),
        x = 'AGE_CD',
        y = 'USE_CNT',
        hue = 'SEX_CD',
        ax = ax[0],
        alpha = .7,
        palette = ['b','skyblue']
    )
    
    sns.barplot(
        data = card_consume_in_DAF.loc[card_consume_in_DAF.MCT_CAT_NM==cat].groupby(['AGE_CD','SEX_CD'])['USE_CNT'].sum().reset_index(),
        x = 'AGE_CD',
        y = 'USE_CNT',
        hue = 'SEX_CD',
        ax = ax[1],
        alpha = .7,
        palette = ['green','yellow']
    )
    
    plt.suptitle(f'{cat}')
    plt.show()
```
